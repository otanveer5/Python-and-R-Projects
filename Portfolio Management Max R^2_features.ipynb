# Step 1

```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'data.csv')

df.next_ret = pd.to_numeric(df.next_ret,errors='coerce')
df = df[~np.isnan(df.next_ret)].reset_index(drop=True)

df['divyield'] = df['divyield'].str.replace('%', '').astype('float')

df.columns = [name.lower() for name in df.columns]

ind_train=df[df.year.isin(range(1980,2000))].index
ind_test=df[df.year.isin(range(2000,2020))].index

df_train=df.loc[ind_train,:].copy().reset_index(drop=True)
df_test=df.loc[ind_test,:].copy().reset_index(drop=True)

df_train
```

# Step 2

```python
feats_not_to_use=["permno","year","month",
"next_ret","pe_op_dil", "mmt6","divyield"]

feats_to_use=[feat for feat in df.columns if feat
not in feats_not_to_use]

target="next_ret"

feats_to_use
```

# Step 3

```python
from numpy.linalg import inv

def normalize(series):
    return (series-series.mean())/series.std()

def linearRegr (feats):
    
    
    mean=df_train[feats].mean(axis=0)
    df_train[feats] = df_train[feats].fillna(mean)
    
    X=df_train[feats].apply(normalize).values
    X = np.concatenate((np.ones(shape=(X.shape[0],1)),X),1)
    theta = inv(X.transpose().dot(X)).dot(X.transpose()).dot(df_train[target].values)
    return theta
    
    
```

# Step 4

```python
def evaluateLinMod (feats, theta):
    mean = df_test[feats].mean(axis=0)
    df_test[feats] = df_test[feats].fillna(mean)      
    X = df_test[feats].apply(normalize).values
    X = np.concatenate((np.ones(shape=(X.shape[0],1)),X),1)
    y_predict = X.dot(theta)
    
    y_true = df_test[target] 
    R2 = 1 - np.sum((y_true - y_predict)**2) / np.sum(y_true**2)  
    return R2
```

# Step 5

```python
feats_list = []
R2values_list=[]
for col in feats_to_use:
    cur_feats = ['mmt6','divyield']
    cur_feats.append(col)
    theta = linearRegr(cur_feats)
    R2values_list.append(evaluateLinMod(cur_feats, theta))
    feats_list.append(cur_feats)
```

```python
R2values_list
```

# Step 6

```python
df1 = pd.DataFrame()
df1['feats_list'] = feats_list
df1['R2values_list'] = R2values_list

df1.sort_values(by=['R2values_list'], inplace=True, ascending=False)
df1.iloc[:3]
```
